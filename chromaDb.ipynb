```python
from langchain_community.vectorstores import Chroma
from langchain_community.embeddings import HuggingFaceEmbeddings
```

```python
from langchain.document_loaders import DirectoryLoader, TextLoader
```

```python
loader = DirectoryLoader("/content/new_articles/", glob = "./*.txt", loader_cls= TextLoader)
```

```python
document = loader.load()
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
```

```python
text_splitter = RecursiveCharacterTextSplitter(chunk_size = 500, chunk_overlap = 100)
text = text_splitter.split_documents(document)
```

```python
from langchain.vectorstores import Chroma
from langchain.embeddings import HuggingFaceEmbeddings
```

```python
persist_directory = 'db'

embedding = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)

vectordb = Chroma.from_documents(
    documents=text,
    embedding=embedding,
    persist_directory=persist_directory
)
```

```python
vectordb.persist()
vectordb = None
```

```python
from langchain.chains import RetrievalQA
```

```python
from langchain.llms import HuggingFacePipeline

from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline
```

```python
model_name = "distilbert/distilgpt2"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name, device_map="auto")
pipe = pipeline("text-generation", model=model, tokenizer=tokenizer, max_new_tokens=200, pad_token_id=tokenizer.eos_token_id )

llm = HuggingFacePipeline(pipeline=pipe)
```

```python
qa_chain = RetrievalQA.from_chain_type(
    llm=llm,
    chain_type="stuff",
    retriever=retriever,
    return_source_documents=True
)
```

```python
def process_llm_response(llm_response):
    print(llm_response['result'])
    print('\n\nSources:')
    for source in llm_response["source_documents"]:
        print(source.metadata['source'])
```

```python
query = "How much money did Microsoft raise?"
llm_response = qa_chain.invoke(query)

process_llm_response(llm_response)
```

```python
query = "What is the news about Pando?"
llm_response = qa_chain(query)
process_llm_response(llm_response)
```

```python
query = "What recent developments have happened with Pando?"
llm_response = qa_chain(query)
process_llm_response(llm_response)
```
